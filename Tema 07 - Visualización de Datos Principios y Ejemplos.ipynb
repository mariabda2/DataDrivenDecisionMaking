##  **Principios de Dise√±o para Visualizaci√≥n Efectiva**

Una visualizaci√≥n bien dise√±ada no solo muestra datos, **los transforma en una historia visual clara e impactante**. Estos principios te ayudar√°n a comunicar con poder y precisi√≥n:

---

## üéØ Claridad**

> Elimina el ruido visual y utiliza t√≠tulos, etiquetas y escalas claras para facilitar la comprensi√≥n.

- Usa fuentes legibles y colores contrastantes.
- A√±ade leyendas completas y ejes bien definidos.
- Evita decoraciones que no aportan valor.

---

## üé® Consistencia

> Mant√©n estilos y colores coherentes a lo largo de todas tus visualizaciones.

- Usa una paleta uniforme.
- Conserva el tipo de gr√°fico por categor√≠a.
- Evita cambios bruscos de formato entre gr√°ficos.

---

## üß† Enfoque

> Dirige la atenci√≥n hacia los datos relevantes utilizando color, forma o tama√±o como herramientas de √©nfasis.

- Resalta valores clave con colores contrastantes.
- Agrega anotaciones para insights importantes.
- Reduce elementos que distraigan.

---

## üß© Proporcionalidad

> Representa los datos respetando sus proporciones reales. Evita escalas enga√±osas o truncadas.

- Verifica que √°reas, tama√±os y longitudes sean correctas.
- Mant√©n escalas coherentes entre gr√°ficos relacionados.
- Evita modificar el eje Y sin justificaci√≥n.

---

## üëÅÔ∏è Intuici√≥n visual

> Usa tipos de gr√°ficos adecuados que faciliten la comprensi√≥n r√°pida.

- Barras para comparar, l√≠neas para tendencias, pastel solo para proporciones simples.
- Ordena las categor√≠as l√≥gicamente.
- Usa convenciones visuales familiares.

---

## üîÑ Reducci√≥n de complejidad

> Descomp√≥n informaci√≥n densa en partes comprensibles.

- Agrupa o categoriza elementos similares.
- Usa capas visuales o dashboards interactivos.
- Presenta datos de forma progresiva si hay mucho contenido.

---

## üìö Narrativa visual

> Cuenta una historia con tus datos: gu√≠a al lector paso a paso.

- Usa t√≠tulos que revelen hallazgos.
- Resalta eventos o anomal√≠as con anotaciones.
- Organiza las visualizaciones en secuencia l√≥gica.

---

## üß≠ Accesibilidad

> Asegura que todos los usuarios puedan interpretar tus gr√°ficos sin barreras.

- Usa colores accesibles (evita rojo/verde conflictivos).
- Incluye texto alternativo y etiquetas claras.
- Prioriza formatos responsivos para distintos dispositivos.

---


## Ejemplos

### üîÑ Reducci√≥n de complejidad

#### ‚ùå Gr√°fico confuso: Demasiadas categor√≠as y etiquetas saturadas

```python
productos = [f"Producto {chr(65+i)}" for i in range(15)]
ventas = [10 * (i+1) for i in range(15)]

plt.figure(figsize=(8, 4))
plt.bar(productos, ventas, color="tomato")
plt.title("Ventas por producto")
plt.xticks(rotation=45)
plt.ylabel("Miles de USD")
plt.tight_layout()
plt.show()
```

#### ‚úÖ Gr√°fico claro: Agrupaci√≥n simplificada para facilitar lectura

```python
import matplotlib.pyplot as plt

categorias = ["Producto A", "Producto B", "Producto C", "Otros"]
ventas = [120, 100, 90, 60]  # 'Otros' agrupa productos menos relevantes

plt.figure(figsize=(6, 4))
plt.bar(categorias, ventas, color="dodgerblue")
plt.title("Ventas por categor√≠a simplificada")
plt.ylabel("Miles de USD")
plt.tight_layout()
plt.show()
```

### üìö Narrativa visual

#### ‚ùå Gr√°fico confuso: datos sin historia ni contexto

```python
plt.figure(figsize=(8, 4))
plt.plot([2020, 2021, 2022, 2023, 2024], [100, 120, 160, 220, 210])
plt.title("Ventas")
plt.tight_layout()
plt.show()
```

#### ‚úÖ Gr√°fico claro: narrativa clara en pasos

```python
import matplotlib.pyplot as plt

# Evoluci√≥n de ventas con evento destacado
a√±os = [2020, 2021, 2022, 2023, 2024]
ventas = [100, 120, 160, 220, 210]

plt.figure(figsize=(8, 4))
plt.plot(a√±os, ventas, marker='o', color="steelblue")
plt.title("Crecimiento de ventas tras el lanzamiento del producto X")
plt.ylabel("Miles de USD")
plt.xlabel("A√±o")
plt.axvline(2021, color="gray", linestyle="--")
plt.text(2021.2, 180, "Lanzamiento de producto X", color="darkred")
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### ‚úÖ Gr√°fico claro: uso de colores accesibles y texto visible
python

```python
import matplotlib.pyplot as plt

categorias = ["Aceptado", "Pendiente", "Rechazado"]
resultados = [60, 25, 15]
colores = ["#0066CC", "#999999", "#CC3300"]  # Altos contrastes y colores distintos para dalt√≥nicos

plt.figure(figsize=(6, 4))
barras = plt.bar(categorias, resultados, color=colores)

# A√±adir etiquetas visibles
for i, valor in enumerate(resultados):
    plt.text(i, valor + 2, f"{valor}%", ha='center', color='black')

plt.title("Estado de solicitudes")
plt.ylabel("Porcentaje")
plt.tight_layout()
plt.show()
```


### üéØ Claridad

#### ‚ùå Gr√°fico confuso: Participaci√≥n de mercado con pastel sobrecargado

```python
import matplotlib.pyplot as plt

# Datos simulados
empresas = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L']
participacion = [12, 10, 9, 8, 7, 6, 6, 6, 6, 5, 5, 4]

# Colores chillones
colores = ['red', 'yellow', 'lime', 'cyan', 'magenta', 'orange', 'purple', 'teal', 'blue', 'pink', 'green', 'brown']

# Gr√°fico circular sin etiquetas adecuadas
plt.figure(figsize=(6, 6))
plt.pie(participacion, colors=colores)
plt.title("Participaci√≥n de Mercado (2025)")
plt.show()
```

#### ‚úÖ Gr√°fico claro: Ventas trimestrales por producto

```python
import matplotlib.pyplot as plt

# Datos simulados
productos = ["Producto A", "Producto B", "Producto C"]
trimestres = ["Q1", "Q2", "Q3", "Q4"]
ventas = [[2.5, 3.0, 3.8, 4.2],  # Producto A
          [1.8, 2.2, 2.9, 3.1],  # Producto B
          [1.2, 1.5, 1.7, 2.0]]  # Producto C

# Crear gr√°fico
plt.figure(figsize=(8, 5))
for i, producto in enumerate(productos):
    plt.plot(trimestres, ventas[i], label=producto, marker="o")

plt.title("Ventas Trimestrales por Producto (2025)")
plt.xlabel("Trimestre")
plt.ylabel("Ventas (Millones USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## üé® Consistencia

#### ‚ùå Gr√°fico confuso: gr√°ficos inconsistentes y desconectados

```python
# Datos alternativos y estilos desordenados
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
ingresos_2024 = [12, 15, 14, 16, 18, 20]
ingresos_2025 = [14, 17, 15, 19, 21, 23]

# Gr√°fico 1 con un estilo
plt.figure(figsize=(8, 4))
plt.bar(meses, ingresos_2024, color='orange')
plt.title("2024 INCOME", fontsize=10)
plt.xlabel("MONTH")
plt.ylabel("USD")
plt.tight_layout()
plt.show()

# Gr√°fico 2 con otro estilo completamente diferente
plt.figure(figsize=(6, 3))
plt.plot(meses, ingresos_2025, linestyle='--', marker='s', color='blue')
plt.title("Ingresos 2025", fontsize=16)
plt.xlabel("Mes")
plt.ylabel("Miles de USD")
plt.tight_layout()
plt.show()
```

#### ‚úÖ Gr√°fico claro: gr√°ficos consistentes de ingresos mensuales

```python
import matplotlib.pyplot as plt

# Datos simulados
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
ingresos_2024 = [12, 15, 14, 16, 18, 20]
ingresos_2025 = [14, 17, 15, 19, 21, 23]

# Gr√°fico 1 (2024)
plt.figure(figsize=(10, 4))
plt.plot(meses, ingresos_2024, marker='o', color='teal')
plt.title("Ingresos Mensuales 2024", fontsize=14)
plt.xlabel("Mes", fontsize=12)
plt.ylabel("Ingresos (mil USD)", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# Gr√°fico 2 (2025), con el mismo estilo
plt.figure(figsize=(10, 4))
plt.plot(meses, ingresos_2025, marker='o', color='teal')
plt.title("Ingresos Mensuales 2025", fontsize=14)
plt.xlabel("Mes", fontsize=12)
plt.ylabel("Ingresos (mil USD)", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## üß† Enfoque

#### ‚ùå Gr√°fico confuso: sin enfoque visual

```python
import matplotlib.pyplot as plt

meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
ventas = [120, 150, 170, 210, 180, 160]

plt.figure(figsize=(8, 4))
barras = plt.bar(meses, ventas, color="lightgray")

# Resaltar el mes con mayor venta
indice_max = ventas.index(max(ventas))
barras[indice_max].set_color("orange")
plt.figure(figsize=(8, 4))
plt.bar(meses, ventas, color=["red", "green", "blue", "purple", "cyan", "yellow"])
plt.title("Ventas 2025")
plt.tight_layout()
plt.show()
```

#### ‚úÖ Gr√°fico claro: Resaltar el mes con m√°s ventas

```python
import matplotlib.pyplot as plt

meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
ventas = [120, 150, 170, 210, 180, 160]

plt.figure(figsize=(8, 4))
barras = plt.bar(meses, ventas, color="lightgray")

# Resaltar el mes con mayor venta
indice_max = ventas.index(max(ventas))
barras[indice_max].set_color("orange")

plt.title("Ventas Mensuales - 2025")
plt.xlabel("Mes")
plt.ylabel("Ventas (mil USD)")
plt.text(meses[indice_max], ventas[indice_max]+5, "Mayor venta", ha='center', color="orange")
plt.tight_layout()
plt.show()
```

#### ‚úÖ Gr√°fico claro: gr√°fico con √≠conos y dise√±o familiar

```python
import matplotlib.pyplot as plt

categorias = ["M√≥vil", "Escritorio", "Tablet"]
usuarios = [65, 25, 10]

plt.figure(figsize=(6, 4))
colores = ["#4CAF50", "#2196F3", "#FFC107"]
barras = plt.barh(categorias, usuarios, color=colores)
plt.title("Origen de usuarios web")
plt.xlabel("% del total")

# Agregar anotaciones
for i, v in enumerate(usuarios):
    plt.text(v + 1, i, f"{v}%", va='center')

plt.tight_layout()
plt.show()
```

## Buenas pr√°cticas

- **Menos es m√°s:** simplifica siempre que puedas.
- **Color con prop√≥sito:** no lo uses solo como decoraci√≥n.
- **Anotaciones √∫tiles:** usa etiquetas breves pero informativas.
- **Contexto:** incluye unidades, escalas, periodos y fuente.

---

## Errores comunes a evitar

- Colores innecesarios o confusos.
- Gr√°ficos 3D innecesarios.
- Ejes sin etiquetas.
- Escalas que distorsionan la informaci√≥n.

---

Una visualizaci√≥n bien dise√±ada permite explorar, comunicar e influir. No se trata solo de mostrar datos, sino de **contar una historia con ellos**.
